##### Context
The purpose of the work is to classify between three different situations in the way people communicate with each other. The first is spontaneous (autonomous) situation in which two people move their hands freely in front of each other.
The second is a synchronous movement in which the two people move their hands together and the third is a movement in position
Own. Where only one side moves the hands.
The idea is to look at the patterns of the hands and try to deduce from them whether it is a situation alone, spontaneous or synchronous.

##### Main Question
Are the in the given input are in synchronized mode?

## Data Proccecing

Our data is very raw, so we will need some proccessing before we can work on it.

```python
right_hand_data.columns = right_hand_data.columns.str.strip() # remove spaces from columns names
```

```python
right_hand_data["Hand Type"].value_counts()
```

First 7 seconds of the recordings does not reflect useful data, so we will get rid of them.

```python
start_time = right_hand_data.head(1)["Time"][0]
right_hand_data.drop(
    right_hand_data[right_hand_data.Time < start_time + 7].index, inplace=True
)
right_hand_data
```

We can see we have some null values:

```python
right_hand_data.isnull().sum(axis = 0)
```

Lets deal with them:

```python
right_hand_data.dropna(inplace=True)
right_hand_data.isnull().sum(axis = 0)
```

We can see we have some left hand values:

```python
right_hand_data["Hand Type"].value_counts()
```

Lets deal with them:

```python
right_hand_data.drop(
    right_hand_data[right_hand_data["Hand Type"] == "left"].index, inplace=True
)
```

```python
right_hand_data["Hand Type"].value_counts()
```

The 'nuber of hands' and 'frame id' columns are useless now so lets drop it too:

```python
right_hand_data.drop(["# hands","Frame ID", "Hand Type"], axis=1, inplace=True)
right_hand_data.reset_index(drop=True)
right_hand_data
```

Lets do the same proccess for the left hand:

```python
left_hand_data.columns = left_hand_data.columns.str.strip() # remove spaces from columns names
start_time = left_hand_data.head(1)["Time"][0]
left_hand_data.drop(
    left_hand_data[left_hand_data.Time < start_time + 7].index, inplace=True
)
left_hand_data.dropna(inplace=True)
left_hand_data.drop(
    left_hand_data[left_hand_data["Hand Type"] == "right"].index, inplace=True
)
left_hand_data.drop(["# hands","Frame ID", "Hand Type"], axis=1, inplace=True)
left_hand_data.reset_index(drop=True)
```

And get their shape equal:

```python
right_hand_data.reset_index(drop=True, inplace=True)
right_hand_data
```

```python
left_hand_data.reset_index(drop=True, inplace=True)
left_hand_data
```

```python
left_hand_data.drop([i for i in range(3932, 4078)], inplace=True)
left_hand_data
```

And finally lets get their times equal:

```python
right_hand_data['Time'] = right_hand_data['Time'] + (left_hand_data.iloc[0,0] - right_hand_data.iloc[0,0])
right_hand_data
```

Now, for all hands we would like to use the 'Time' column as our index:

```python
right_hand_data = right_hand_data.set_index('Time')
left_hand_data = left_hand_data.set_index('Time')
sync_data = sync_data.set_index('Time')
spontan_data = spontan_data.set_index('Time')
```
